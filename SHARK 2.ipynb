Escolha dos primeiros 10 paises da Oceania com mais ataques de tubarão

```python
df4 = df3.drop(df3[(df3.country != 'AUSTRALIA') & (df3.country != 'PAPUA NEW GUINEA') & (df3.country != 'NEW ZEALAND') & (df3.country != 'FIJI') &(df3.country != 'NEW CALEDONIA') & (df3.country != 'SOLOMON ISLANDS') & (df3.country != 'FRENCH POLYNESIA') & (df3.country != 'TONGA') & (df3.country != 'VANUATU') & (df3.country != 'MARSHALL ISLANDS')].index)
```

```python
oceania = pd.DataFrame(df4.country.value_counts().reset_index().values, columns=["country", "attacks"])
oceania = oceania.sort_index(axis = 0, ascending=True)
oceania
```

Coluna nova km de costa de cada país (enchendo linguiça)

```python
coastline = [34218, 5150, 15000, 1129, 2254, 5313, 2525, 419, 2528, 370]

oceania['coastline_km'] = coastline
```

```python
oceania
```

```python
oceania.info()
```

```python
oceania['attacks'] = oceania['attacks'].astype(str).astype(float)
oceania['coastline_km'] = oceania['coastline_km'].astype(str).astype(float)
```

```python
oceania.info()
```

Nova coluna ataques de tubarão por km de costa

```python
oceania['attacks/km'] = oceania['attacks'] / oceania['coastline_km']
```
